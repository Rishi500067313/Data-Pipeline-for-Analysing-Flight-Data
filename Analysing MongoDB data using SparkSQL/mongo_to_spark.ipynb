```
spark.sql("select distinct(flight_number), depAirport, arrAirport from ind_fli where arrTimezone like 'Asia%'").show()
```

```
spark.sql("select distinct(arrTimezone) from ind_fli").show(300)
```

```
spark.sql("""select distinct(flight_number),depAirport as Departure, arrAirport as Arrival, Rank , (arrDelay+depDelay) as Total_delay 
                from (select flight_number, depAirport, arrAirport, arrDelay, depDelay, rank() over (order by (arrDelay+depDelay) desc) as Rank from ind_fli where (arrDelay+depDelay) is not null)""").show()
```

```
spark.sql("select distinct(airlineName) as Name_of_Airline, speedhori as Speed from ind_fli where speedhori > 500 order by speed desc").show()
```

```
spark.sql("select distinct flight_number, airlineName as Airlines, altitude from ind_fli order by altitude desc limit 5").show()
```
